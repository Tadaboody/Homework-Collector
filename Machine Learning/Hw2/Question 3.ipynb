We initialize our data:

```python
from scipy.io import loadmat
from parzen_window import *
import matplotlib.pyplot as plt
mat = loadmat("dataAB.mat")
a_train, b_train, a_val, b_val, a_test, b_test = load_data(
    mat, 'train_dataA', 'train_dataB', 'valid_dataA', 'valid_dataB', 'test_dataA', 'test_dataB')
clf = PNN(['a', 'b'])
clf.fit({'a': a_train, 'b': b_train}, 1)
```

```python
opt_window, window_sizes, success_rates = find_optimal_window_size(
    clf, a_val, b_val)
fig, ax = plt.subplots()
ax.plot(window_sizes,success_rates,'-')
plt.title("Success rate of validation set over window size")
plt.show()
print("optimal window size is: " + str(opt_window))
```

```python
success_rate = classify_data(clf, opt_window, a_test, b_test)
print("success rate on the test set is {}%".format(str(success_rate)))
```
